# 2.1.5. Saving Memory

```python
before = id(Y)
Y = X + Y
after = id(Y)
before == after
```

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

```python
before = id(X)
X += Y
after = id(X)
before == after
```
