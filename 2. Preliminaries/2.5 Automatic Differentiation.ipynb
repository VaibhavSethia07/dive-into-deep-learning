# 2.5.1. A Simple Function

```python
import torch
```

```python
x = torch.arange(4.0)
x
```

```python
# Can also create x = torch.arange(4.0, requires_grad=True)
x.requires_grad_(True)
x.grad  # The gradient is None by default
```

```python
y = 2*torch.dot(x, x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad == 4*x
```

```python
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```

# 2.5.2. Backward for Non-Scalar Variables

```python
x.grad.zero_()
y = x*x
y.backward(gradient=torch.ones(len(y)))
x.grad
```

# 2.5.3. Detaching Computation

```python
x.grad.zero_()
y = x * x
u = y.detach()
z = u * x

z.sum().backward()
x.grad
```

```python
x.grad == u
```

# 2.5.4. Gradients and Python Control Flow

```python
def f(a):
    b = a * 2
    while b.norm() < 1000:
        b = b * 2
    if b.sum() > 0:
        c = b
    else:
        c = 100 * b
    return c
```

```python
a = torch.randn(size=(), requires_grad=True)
d = f(a)
d.backward()
```

```python
a.grad == d / a
```
